Logistic regression 

Recall, initialize w and b

Predict y_hat = sigmoid(xw + b)

Calculate the loss

loss.backward() to get gradients

Update weight accordingly

```python
w = torch.randn((2,2), requires_grad = True)
```

```python
b = torch.randn((1,2), requires_grad = True)
```

CrossEntropyLoss has softmax inside of it

```python
y_hat = x @ w + b
```

```python
CrossEntropy = torch.nn.CrossEntropyLoss()
```

```python
loss = CrossEntropy(y_hat, torch.argmax(y, dim = 1))
```

```python
loss.backward()
```

```python
w.grad
```

```python
lr = 0.01
```

```python
w = w - lr * w.grad
```

```python
w = w.detach() # removes from the computational graph
```

```python
w.requires_grad = True
```

```python
b = b - lr * b.grad
```

```python
b = b.detach()
```

```python
b.requires_grad = True
```

In a for loop to train our logistic regression model

```python
lr = 0.001
iterations = 10000

for i in range(0, iterations):
    y_hat = x @ w + b
    CrossEntropy = torch.nn.CrossEntropyLoss()
    loss = CrossEntropy(y_hat, torch.argmax(y, dim = 1))
    loss.backward()
    w = w - lr * w.grad
    w = w.detach()
    w.requires_grad = True
    b = b - lr * b.grad
    b = b.detach()
    b.requires_grad = True    
    print(loss)
```

You can also use optimizers to not have to detach()/reattach all the time

```python
class Layer(torch.nn.Module):
    def __init__(self, size_in, size_out, activation_func):
        super().__init__()
        self.w = torch.nn.Parameter(torch.randn(size_in, size_out, requires_grad = True))
        self.b = torch.nn.Parameter(torch.randn(1, size_out, requires_grad = True))
        self.activation_func = activation_func
    
    def forward(self, z_in):
        return self.activation_func(z_in @ self.w + self.b)
```

```python
model = Layer(2, 2, activation_func = torch.sigmoid)
```
